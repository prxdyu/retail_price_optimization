```python
# importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# turning off warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# importing the sales data
df = pd.read_csv("data/data.csv")
df.head(5)
```

### Removing Effects

Now this df consists of transactions made in all days but we know that  on holidays,weekend and school breaks the price of the food items change, so we are going to remove transactions made on those dates to remove the effect due to the holidays,weekend and school breaks so that we can determine the typical price of the product on the usual day

```python
# removing transactions made on Holidays, Week-ends, School breaks
df_filtered = df[(df['HOLIDAY'] == "No Holiday") & (df['IS_SCHOOLBREAK'] == 0) & (df['IS_WEEKEND'] == 0)]
```

#### 1. Food items

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='ITEM_NAME', data=df, palette='viridis')

# Add titles and labels
plt.title('Item Count Plot')
plt.xlabel('Item Name')
plt.ylabel('Count')

# Show the plot
plt.show()
```

We can see that most of the people buy burger followed by coke, lemonade and coffee

#### 2. Distribution of Price

```python

# Create subplots with two plots side by side
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Plot the first Kernel Density plot
sns.kdeplot(df['PRICE'], shade=True, color='b', ax=axes[0])

# Add titles and labels to the first plot
axes[0].set_title('Kernel Density Plot of Price (Original')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Density')

# Plot the second Kernel Density plot
sns.kdeplot(df_filtered['PRICE'], shade=True, color='b', ax=axes[1])

# Add titles and labels to the second plot
axes[1].set_title('Kernel Density Plot of Price (Filtered)')
axes[1].set_xlabel('Price')
axes[1].set_ylabel('Density')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

From the KDE plot we can clearly see that there are more than single peaks which means it's bi-modal distribution

#### 3. Price  vs Quantity

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(9, 4))
sns.scatterplot(x='PRICE', y='QUANTITY', data=df)

# Add titles and labels
plt.title('Scatter Plot of Price vs Quantity')
plt.xlabel('Price')
plt.ylabel('Quantity')

# Show the plot
plt.show()
```

We can clearly see that there are two groups in the scatter plot

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(9, 4))
sns.scatterplot(x='PRICE', y='QUANTITY', hue="ITEM_NAME",data=df)

# Add titles and labels
plt.title('Scatter Plot of Price vs Quantity')
plt.xlabel('Price')
plt.ylabel('Quantity')

# Show the plot
plt.show()
```

#### 4. Pair Plot

```python
# Create the count plot with increased size
plt.figure(figsize=(12, 9))
sns.pairplot(df[["PRICE","QUANTITY","ITEM_NAME"]],hue="ITEM_NAME",plot_kws={'alpha':0.1})
plt.show()
```

We can see the group in the top right corner of the scatter plot is burgers
